# Constraints

1. Constraints r rules enforced on data columns on table.
2. They r used to prevent invalid data from being entered into database.
3. This ensures accuracy & reliability of data in database.

## Types of Constraints

### 1. Column Constraint
applied to individual column


### 2. Table Constraint
applied to entire table

## Most Common Column Constraints used -



1. **Not Null Constraint** - a column can't have null value. e.g. email


2. **Unique Constraint** - all values in column r different. e.g. customer_id


3. **Primary Constraint** - uniquely identify each row/record


4. **Foreign Constraint** - constraints data based on columns in other tables. When u set up a foreign key u set up a relationship to another table.


5. **Check Constraint** - all values satisfy some condition e.g. has to be less than 20.


6. **Exclusion Constraint**



## Most Common Table Constraints used -


1. **Check Constraint**
2. **References Constraint**
3. **Unique(column_list)** - allows to expand unique key constraint to many columns
4. **Primary Key(column_list)** - allows to define primary key that consists of multiple columns.
