# Data Types

## Documentation-

https://www.postgresql.org/docs/current/datatype.html

When u need to choose data types for creating your tables and databases, u can refer to available options from above documentation.

e.g. when u think u need to choose numeric data type u might have below options for numeric. Similarly many options for each data type.

In broad category, below data types and each of them have several options to choose from as for numeric data type above.

**1. Boolean** - True or False

**2. Character** - char, varchar, text

**3. Numeric** - integer and floating point number

**4.Temporal** - date, time, timestamp, and interval.

Other important ones-

**5. UUID** -  Universally Unique Identifiers

**6. Array** - Stores an array of strings, numbers etc

**7. JSON**

**8. Hstore key-value pair**

**9. Special types such as network address and geometric data**

SERIAL DATA TYPE to create automatically unique integers (unique identifiers). If row is later removed SERIAL data type will not adjust, marking fact that row was removed from sequence. You will find serial under numeric data types.
